```python
def exportar_tablas_a_csv():
    print("🚀 Iniciando exportación total de la arquitectura...")

    # 1. Exportar Metadatos (La configuración y auditoría)
    print("\n📜 Exportando Capa de Metadatos...")
    try:
        query_meta = text("SELECT * FROM meta.pipeline_config")
        df_meta = pd.read_sql(query_meta, engine)
        df_meta.to_csv("meta_pipeline_config.csv", index=False)
        print("   ✅ Generado: meta_pipeline_config.csv")
    except Exception as e:
        print(f"   ❌ Error en Metadatos: {e}")

    # 2. Exportar Capa Silver (Data Limpia)
    tablas_silver = [
        'sap_clientes', 'sap_productos', 'sap_canales', 'ads_campanas',
        'sap_ventas_cabecera', 'sap_ventas_detalle', 'sap_inventario_diario',
        'ads_insights_diario', 'clima_diario_log'
    ]
    print("\n📂 Exportando Capa Silver...")
    for tabla in tablas_silver:
        try:
            query = text(f"SELECT * FROM silver.{tabla}")
            df = pd.read_sql(query, engine)
            df.to_csv(f"silver_{tabla}.csv", index=False)
            print(f"   ✅ Generado: silver_{tabla}.csv")
        except Exception as e:
            print(f"   ❌ Error en silver.{tabla}: {e}")

    # 3. Exportar Capa Gold (Dataset Maestro para ML)
    print("\n🏆 Exportando Capa Gold...")
    try:
        query_gold = text("SELECT * FROM gold.ml_feature_set")
        df_gold = pd.read_sql(query_gold, engine)
        df_gold.to_csv("gold_ml_feature_set.csv", index=False)
        print("   ✅ Generado: gold_ml_feature_set.csv")
    except Exception as e:
        print(f"   ❌ Error en Gold: {e}")

    print("\n🏁 Proceso terminado. ¡Ya tienes el mapa completo de tu base de datos!")

# Ejecutar
exportar_tablas_a_csv()
```
